```python
# ðŸŽµ SoundFont File Path (Update this path!)
# SOUNDFONT_PATH = "C:\\SoundFonts\\FluidR3_GM.sf2"  # Replace with your actual path
import fluidsynth
import time
SOUNDFONT_PATH="C:\\Users\\vedha\\Desktop\\Music clg\\GeneralUser_GS_v2.0.1--doc_r3\\GeneralUser-GS\\GeneralUser-GS.sf2"
# Initialize FluidSynth
fs = fluidsynth.Synth()
fs.start()
sfid = fs.sfload(SOUNDFONT_PATH)  # Load SoundFont
fs.program_select(0, sfid, 0, 0)  # Select Instrument (0 = Piano)

# ðŸŽ¶ Define Notes for Each Button
base_notes = {
    "button1": 60, "button2": 62, "button3": 64, "button4": 65,
    "button5": 67, "button6": 69, "button7": 71, "button8": 72
}

# ðŸŽ¶ Define Gyroscope Effects
gyro_effect = {
    "gyro_left": -2,   # Decrease pitch
    "gyro_right": +2   # Increase pitch
}

# ðŸŽ¶ Define Accelerometer Effects
accel_effect = {
    "accel_forward": +1,   # Slightly higher
    "accel_backward": -1   # Slightly lower
}

# ðŸŽ¹ Generate Full Music Map
music_map = {}
for button, note in base_notes.items():
    for gyro, g_effect in gyro_effect.items():
        for accel, a_effect in accel_effect.items():
            final_note = note + g_effect + a_effect
            music_map[f"{button},{gyro},{accel}"] = [final_note]

# ðŸŽµ Function to Play Music Based on Sensor Input
def play_music(input_combination):
    notes = music_map.get(input_combination.strip(), None)  # Get MIDI notes
    if notes:
        print(f"ðŸŽ¶ Playing Notes: {notes}")
        for note in notes:
            fs.noteon(0, note, 100)  # Play note
        time.sleep(0.5)  # Hold notes
        for note in notes:
            fs.noteoff(0, note)  # Stop note
    else:
        print(f"âš  Unknown input: {input_combination}")

# ðŸ”„ Main Loop: Simulate Sensor Input
while True:
    # Simulate receiving input from ESP32 sensor data
    input_combination = input("Enter sensor data (button, gyro, accel) or 'exit' to quit: ")
    if input_combination.lower() == 'exit':
        print("Exiting...")
        break
    play_music(input_combination)
```
